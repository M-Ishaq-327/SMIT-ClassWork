```python
import pandas as pd
```

```python
df = pd.read_csv('nobel.csv')
df.head()
```

```python
df['birth_date']= pd.to_datetime(df['birth_date'],errors= "coerce")
```

```python
df['born_week_day']= df['birth_date'].dt.day_name()
df.head()
```

### Group By

```python
#find total count of nobel prize.
df.groupby(['category'])['prize'].count()
```

```python
import pandas as pd
df_tiatnic = pd.read_excel('titanic.xlsx')
df_tiatnic.head()
```

```python
df_tiatnic.groupby('sex')[['fare','age']].mean()
```

```python
df_tiatnic.groupby('parch')['age'].agg(['mean','median','max','min'])
```

Mapping function


```python
df_tiatnic['sex'] = df['sex'].map({'Male':'M','Female':'F'})
df_tiatnic['sex'].head(20)
```

APPLY FUNCTION


```python
#nobel file 
df['prize_year_cat']=df['year'].apply(lambda x: '20th century' if x<2000 else '21st century')
df['prize_year_cat']
```

### ApplyMap Function

```python
df[["year","laureate_id"]].applymap(lambda x: x+1)
```

Pivot Table

```python
pivot = df_tiatnic.pivot_table(values='fare',index='parch',columns='survived',aggfunc='mean')
pivot
```

# Crosstab

```python
crosstab = pd.crosstab(df_tiatnic['pclass'],df_tiatnic['sex'],values=df_tiatnic['survived'],aggfunc='mean')
crosstab
```
